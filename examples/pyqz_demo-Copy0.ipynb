# Introducing the new pyqz

The pyqz Python module was first publicly released as <b>pyqz 0.4</b> in <b>Dopita+(2013)</b>.

The code was later on improved to propagate observational errors, via the reconstruction of the 2-D probability density function in the {log(Q) vs 12+log(O/H)} plane in <b>pyqz v0.6</b>. The updated code was not publicly released, but described in Appendix B of <b>Vogt+(2015)</b>, and has been available on dedmand since then.

Recent heroic improvements to the MAPPINGS code have warranted further revisions and upgrades of pyqz. This document briefly explores the different capabilites of the code with basic examples to get you started quickly. 

<b><i>Warning: the code syntax has changed significantly, and so did the function calls. pyqz v0.7.0 is therefore NOT backward compatible with previous pyqz versions.</i></b>

### 3) A note on the pyqz syntax

The pyqz module is intimately linked to the MAPPIINGS code. While both are stand alone and distinct programs, pyqz was designed to employ the same notation conventions that that of the MAPPINGS code for clarity, both from a user and programming perspective.

These syntax conventions are as follows:

- the ionization parameter is <code>'logQ'</code>,
- the oxygen abundance is <code>'Tot[O]+12'</code> (total) or <code>'gas[O]+12'</code> (for the gas-phase),
- the Balmer lines from Hydrogen are <code>'Ha'</code>, <code>'Hb'</code>, etc ...,
- forbidden lines are marked as <code>'[OIII]'</code>, <code>'[NII]'</code>, <code>'[SII]'</code>, <code>'[OI]'</code>, etc ..., and
- for the usual strong line doublets, when the doublet line fluxes are considered together (e.g. [OIII]5007 + 4959), a <code>'+'</code> is appended to the said emission line (e.g. <code>'[OIII]+'</code>). By convention, the single line is always the strongest within the doublet. In short, <code>'[OIII]'</code> corresponds to [OIII]5007, <code>'[OIII]+'</code> corresponds to [OIII]5007+4959, <code>'[NII]'</code> corresponds to [NII]6584, <code>'[NII]+'</code> corresponds to [NII]6584+6548, etc ...

This syntax must be followed carefully when using pyqz, or errors will arise.

### 4) How to use pyqz: basic examples

pyqz gives you access to the latest MAPPINGS strong line ratio diagnostic diagrams (and associated info on the MAPPINGS version used to generate the grid, etc...). You can access the nodes of any diagram using:

```python
pyqz.get_grid('[NII]/[SII]+;[OIII]/[SII]+', struct='pp', sampling=1)
```

Alternatively, you may wish to simply explore how the diagnostic diagram looks (and e.g. check whether the MAPPINGS grid is flat, or wrapped). An important new feature of pyqz is the auto-detection of wraps in the simulation grid, marked with red segments in the diagram. For example, try to look at '[NII]/[OII]+;[OIII]/[SII]+'.

```python
pyqz.check_grid('[NII]/[OII]+;[OIII]/[SII]+',struct='pp',show_plot=True, sampling=1)
```

By default, the MAPPINGS grid are corse. For various reasons better explained elsewhere (see the MAPPINGS documentation), only a few points have matching stellar tracks <b>AND</b> stellar atmospheres (in terms of abundances). Hence, only a few abundance points can be simulated in a consistent fashion. However, pyqz contains a <b>resampled</b> version of each grid (constructed using <code>pyqz.refine_MVphotogrid()</code>, see the docs for more info), which you can use. The resampling is performed in the {log(Q) or 12+log(O/H) vs line ratio} space for all line ratios returned by MAPPINGS using Akima splines. For example:

```python
pyqz.check_grid('[NII]/[SII]+;[OIII]/[SII]+',struct='pp',show_plot=True, sampling=2)
```

At the core of pyqz lies <code>pyqz.interp_qz()</code>, which is the basic routine used to interpolate a given line ratio diagnostic grid. The function is being fed by line ratios stored inside numpy arrays, and will only return a value for line ratios landing <b> on valid and un-wrapped</b> regions of the grid:

```python
pyqz.interp_qz('Tot[O]+12',[np.array([0.0,-0.8]),np.array([-0.2,-0.2])],'[NII]/[OII]+;[OIII]/[SII]+', 
               sampling=1,struct='pp', show_plot=True)
```

Of course, one usually wants to compute both log(Q) and 12+log(O/H) for a large set of strong emission line fluxes, and combining the estimates from different line ratio diagnostics diagrams introduced by Dopita+(2013). For existing Python users, this is exactly what the function <code>pyqz.get_global_qz()</code> allows you to do from within your existing Python script. The function is being fed the data directly in the form of numpy arrays and lists. Less experienced Python users may instead be interested in feeding all this information via an input csv file (and in turn be given another csv file with the results), which is exactly what the function <code>pyqz.get_global_qz_ff()</code> does. 

```python
pyqz.get_global_qz(np.array([[2.88,0.11,1.0,0.05,2.015,0.045,0.036,0.002,2.72,0.12,0.497,0.017,0.500,0.016]]),
                   ['[OII]+','std[OII]+','Hb','stdHb','[OIII]','std[OIII]','[OI]','std[OI]','Ha','stdHa','[NII]','std[NII]','[SII]+','std[SII]+'], 
                   ['[NII]/[SII]+;[OIII]/[SII]+','[NII]/[OII]+;[OIII]/[OII]+'], 
                   ids = ['NGC_1234'],
                   show_plot=True,
                   #save_plot=True,
                   #plot_loc = './',
                   struct='pp',
                   sampling=1)
```

### 5) Other "advanced" features

pyqz does support 3-D line ratios:

```python
bad_segments = pyqz.check_grid('[NII]/[SII]+;[NII]/Ha;[OIII]/Hb',
                coeffs = [[1.0,0.264,0.0],[0.242,-0.910,0.342]],
                show_plot=True,
                struct='pp',
                sampling=1)
```

pyqz does handle upper limits for the line fluxes (which are indicated with errors=-1):

```python
pyqz.get_global_qz(np.array([[2.015,-1,0.497,0.017,0.500,0.016]]),
                   ['[OIII]','std[OIII]','[NII]','std[NII]','[SII]+','std[SII]+'], 
                   ['[NII]/[SII]+;[OIII]/[SII]+'], 
                   ids = ['NGC_1234'],
                   show_plot=True,
                   #save_plot=True,
                   #plot_loc = './',
                   struct='pp',
                   sampling=1)
```
