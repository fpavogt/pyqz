# Running pyqz: basic examples

### Installing and importing pyqz

Installing pyqz is now easier than ever. Just download it from Github, unpack it anywhere you like, and make sure that location is in your Python path. You should then be able to import the package and check its version:

```python
import pyqz
pyqz.__version__
```

### Basic use 1: accessing MAPPINGS line ratio diagnostic grids

pyqz gives you easy access to the latest MAPPINGS strong line ratio diagnostic diagrams (and associated info on the MAPPINGS version used to generate the grid, etc...). This can for example be useful to create your own line ratio diagnostic plots. You can access the nodes of any line ratio diagram using:

```python
a_grid = pyqz.get_grid('[NII]/[SII]+;[OIII]/[SII]+', struct='pp', sampling=1)
```

Alternatively, you may wish to simply explore how a given diagnostic diagram looks (and e.g. check whether the MAPPINGS grid is flat, or wrapped) for line ratios of your choice. An important feature of pyqz is the auto-detection of wraps in the simulation grid, marked with red segments in the diagram, and returned by the function <i> pyqz.check_grid()</i>:

```python
bad_segments = pyqz.check_grid('[NII]/[OII]+;[OIII]/[SII]+',struct='pp',show_plot=True, sampling=1)
```

By default, the MAPPINGS grid are corse. For various reasons better explained elsewhere (see the MAPPINGS documentation), only a few points have matching stellar tracks <b>AND</b> stellar atmospheres (in terms of abundances). Hence, only a few abundance points can be simulated in a consistent fashion. 

Rather then 1) interpolating between stellar tracks and stellar atmosphere and 2)running extra MAPPINGS models (which would use hence inconsistent input), pyqz contains a <b>resampled</b> version of each grid (constructed using <code>pyqz.refine_MVphotogrid()</code>, see the docs for more info). The resampling is performed in the {<code>logQ</code> and <code>Tot[O+12]</code> vs line ratio} space for all line ratios returned by MAPPINGS using Akima splines. Resampled grids can be accessed via the <i>sampling</i> keyword, e.g.:

```python
bad_segments = pyqz.check_grid('[NII]/[OII]+;[OIII]/[SII]+',struct='pp',show_plot=True, sampling=2)
```

In the default pyqz diagrams, the <b>original</b> MAPPINGS nodes are circled with a black outline, while the <b>reconstructed</b> nodes are not.

Besides the <code>sampling</code> keyword, 
  - <code>Pk</code> let's you define the pressure of the simulated HII regions,
  - <code>struct</code> allows you to choose between plane-parallel (<code>'pp'</code>) and spherical (<code>'sph'</code>) HII regions, and
  - <code>kappa</code> lets you define the value of $\kappa$ (from the so-called $\kappa$-distribution).
  
  Note: currently, the values of <code>Pk</code> and <code>kappa</code> <b> must match existing MAPPINGS models</b> located in <i>pyqz/reference_data/</i>, either provided by default with the code, or computed by the user.

### Basic use 2: deriving <code>logQ</code> and <code>Tot[O+12]</code> for a given set of line ratios

At the core of pyqz lies <code>pyqz.interp_qz()</code>, which is the basic routine used to interpolate a given line ratio diagnostic grid. The function is being fed by line ratios stored inside numpy arrays, and will only return a value for line ratios landing <b> on valid and un-wrapped</b> regions of the grid:

```python
z = pyqz.interp_qz('Tot[O]+12',[np.array([-0.6]),np.array([-0.1])],'[NII]/[OII]+;[OIII]/[SII]+', 
               sampling=1,struct='pp', show_plot=True)
print z
```

Of course, one usually wants to compute both <code>logQ</code> and <code>Tot[O+12]</code> or <code>gas[O+12]</code> for a large set of strong emission line fluxes, combining the estimates from different line ratio diagnostics diagrams. This is exactly what the function <code>pyqz.get_global_qz()</code> allows you to do. 

The function is being fed the individual line fluxes and associated errors in the form of numpy arrays and lists. Identification tages for each dataset can also be given to the function - these are then used when saving the different diagrams.

Less experienced Python users may instead be interested in feeding all this information via an input csv file (and in turn be given another csv file with the results), which is exactly what the function <code>pyqz.get_global_qz_ff()</code> does. 

```python
pyqz.get_global_qz(np.array([[2.88,0.05,1.0,0.01,2.015,0.045,0.036,0.002,2.72,0.12,0.497,0.017,0.500,0.016]]),
                   ['[OII]+','std[OII]+','Hb','stdHb','[OIII]','std[OIII]','[OI]','std[OI]','Ha','stdHa','[NII]','std[NII]','[SII]+','std[SII]+'], 
                   ['[NII]/[SII]+;[OIII]/[SII]+','[NII]/[OII]+;[OIII]/[OII]+'], 
                   ids = ['NGC_1234'],
                   show_plot=True,
                   #save_plot=True,
                   #plot_loc = './',
                   struct='pp',
                   sampling=1)
```

At this time, all line fluxes measurements errors are assumed to be gaussian, except if line fluxes are tagged as upper-limits (which are indicated with errors=-1).

### Advanced usage 1: "3-D" line ratio diagrams

pyqz does support 2-D line ratio diagrams constructed from 3 sets of line ratios (i.e. 3-D line ratio diagrams projected to a given 2-D plane):

```python
bad_segments = pyqz.check_grid('[NII]/[SII]+;[NII]/Ha;[OIII]/Hb',
                coeffs = [[1.0,0.264,0.0],[0.242,-0.910,0.342]],
                show_plot=True,
                struct='pp',
                sampling=1)
```
